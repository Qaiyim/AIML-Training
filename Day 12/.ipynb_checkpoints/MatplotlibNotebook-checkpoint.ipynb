# First example to create a line chart from our data
    ## Students.xlsx for fetching data
    ## Generate Line Chart based on students name and students marks

```python
import pandas as pd
```

```python
df=pd.read_excel('students.xlsx')
```

```python
print(df)
```

```python
Names=df['Name']
```

```python
Marks=df['Marks']
```

```python
print(Names)
print(Marks)
```

```python
plt.figure(figsize=(10,8))
plt.plot(Names,Marks,marker='*',color='red')
plt.title('Students Marks Line Chart')
plt.xlabel('Student Names')
plt.ylabel('Students Marks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(Marks,Names,marker='o',color='green')
plt.title('Students Performance')
plt.xlabel('Student Marks')
plt.ylabel('Students Names')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(Names,Marks,marker='o',color='green',linestyle='--')
plt.title('Students Performance')
plt.xlabel('--Names--')
plt.ylabel('--Marks--')
plt.grid(True)
plt.show()
```

```python
print('Dotted Line')
plt.figure(figsize=(10,8))
plt.plot(Names,Marks,marker='o',color='green',linestyle=':')
plt.title('Students Performance')
plt.xlabel('--Names--')
plt.ylabel('--Marks--')
plt.grid(True)
plt.show()
```

```python
print('Dash Line')
plt.figure(figsize=(10,8))
plt.plot(Names,Marks,marker='o',color='orange',linestyle='-.')
plt.title('Students Performance')
plt.xlabel('--Names--')
plt.ylabel('--Marks--')
plt.grid(True)
plt.show()
```
