# Scatter Plots
    ## In scatter plots we represent data in points, where each point represent a data observation, we can check how to values correlate 

```python
product_sales={'Laptop':750,'Smartphone':650,'Smartwatch':400,'Tablets':500,'Ac':400}
products=list(product_sales.keys())
sales=list(product_sales.values())
plt.scatter(products,sales)
plt.title('Basic Product and Sales Representation')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
players=['Qaiyim','Ronaldo','Messi','Chanatip','Khawin']
match1_score=[90,78,57,89,56]
match2_score=[67,56,34,78,56]
plt.figure(figsize=(8,5))
plt.scatter(players,match1_score,color='red',label='Match One', s=100, marker='*')
plt.scatter(players,match2_score,color='green',label='Match One', s=100, marker='x')
plt.title('Comparison of Players in Two Matches')
plt.xlabel('Players')
plt.ylabel('Score')
plt.grid(True,linestyle='--', alpha=1)
plt.legend(loc='lower left')
plt.show()
```

# Write a comparison that shows average calories burned for each workout type from final_data dataset

```python
df=pd.read_csv('Final_data.csv')
df.head()
print(df['Workout_Type'])
print(df['Calories_Burned'])
avg_calories=df.groupby('Workout_Type')['Calories_Burned'].mean()
print(avg_calories)
plt.figure(figsize=(8,5))
plt.scatter(avg_calories.index,avg_calories.values, color='red',label='Match One', s=200, marker='*')
plt.title('Comparison of Average Calories Burned for Each Workout Type')
plt.xlabel('Workout_Type')
plt.ylabel('Calories_Burned')
plt.grid(True,linestyle='--', alpha=1)
plt.legend(loc='lower left')
plt.show()
```
