# Give the list of patients, whose are in ICU
## Save the file with icu_patients.csv

```python
icu_patients=patients[patients['service']=='ICU']
icu_patients
```

```python
icu_patients.to_csv('Icu_patients.csv',index=False)
print('File created')
```
