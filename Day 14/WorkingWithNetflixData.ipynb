```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('netflix_data.csv')
df.head(3)
```

```python
sns.set_theme(style='dark',palette='coolwarm')
sns.countplot(x='type',data=df, hue='type')
plt.title('Number of Movies vs Tv Shows')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
df=df.dropna(subset=['type','genre'])
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='type',data=df.head(10), hue='country', palette='Set2')
plt.title('Netflix Content By Country')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='type',data=df, hue='country', palette='Set2')
plt.title('Netflix Content By Country')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.legend(title='Country', bbox_to_anchor=(1.02,1),loc='upper left', frameon=True, borderaxespad=0, ncol=1)
plt.tight_layout()
plt.show()
```
