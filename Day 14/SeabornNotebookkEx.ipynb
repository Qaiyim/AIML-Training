# Seaborn Library of Matplotlib
## Seaborn is a data visualization library in Matplotlib that provides high level interface for creating interactive with less code


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
palette=sns.color_palette()
print(palette)
```

```python
sns.palplot(palette)
```

```python
palette_green=sns.color_palette('Greens',30)
sns.palplot(palette_green)
```

```python
data_dict={
    'building':['Burz-Khalifa','KLCC','Alor Setar Tower'],
    'category':['Highest','Medium','Low'],
    'building_height':[150,120,90]}

bar_colors=['maroon','green','purple']
data=pd.DataFrame(data_dict)
sns.barplot(x='building',y='building_height',hue='category',data=data,palette=bar_colors)
plt.show()


```

```python
data_dict={
    'building':['Burz-Khalifa','KLCC','Alor Setar Tower'],
    'category':['Highest','Medium','Low'],
    'building_height':[150,120,90]}

data=pd.DataFrame(data_dict)
sns.barplot(x='building',y='building_height',hue='category',data=data,palette='coolwarm')
plt.show()
```

```python
data_iris=sns.load_dataset('iris')
print(data_iris)
```

```python
sns.barplot(x='species',y='sepal_width',data=data_iris)
plt.show()
```

```python
sns.violinplot(x='species',y='petal_width',data=data_iris,palette=bar_colors)
plt.show()
```

```python
data_iris=sns.load_dataset('iris')
sns.pairplot(hue='species',data=data_iris)
plt.show()
```

```python
sns.set_style('ticks')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('Darkgrid Theme Example Using Iris Data')
plt.show()
```

```python
# white, tick, darkgrid, whitegrid, dark
sns.set_style('whitegrid')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('Darkgrid Theme Example Using Iris Data')
plt.xlabel('Sepal Length in cm')
plt.ylabel('Sepal Width in cm')
plt.xticks([4,5,6,7,8], fontsize=10, color='maroon')
plt.yticks([1,2,3,4,5,6,7], fontsize=10, color='maroon')
sns.despine()
plt.show()
```
