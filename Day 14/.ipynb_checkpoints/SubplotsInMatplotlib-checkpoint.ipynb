# Matplotlib.pyplot.Subplot()
## subplots(): provides a way to plot multiple plots on single figure.
## Given number of rows and columns, it returns a tuple(fig.ax) giving a single figure fig with an array of axes ax

```python
import matplotlib.pyplot as plt
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
```

```python
import numpy as np
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
x=np.random.random(10)
ax1.plot(x,x/2)
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
x=np.random.random(10)
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
```

```python
sales_data={'Electronics':45000,'Clothing':30000,'Grocery':25000,'Mobile':20000}
category=list(sales_data.keys())
actual_sales=list(sales_data.values())
print(category)
print(actual_sales)

fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
x=np.random.random(10)
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(category,actual_sales)
ax4.hist(np.random.randn(1000))
```

# Subplots using indexing

```python
our_line=np.linspace(0,10,100)
```

```python
product_prices={'Hair Oil':20,'Face Wash':25,'Shampoo':14,'Sunscreen':23}
```

```python
our_numbers=np.random.random(10)
```

```python
our_dots=np.random.randn(1000)
```

```python
actual_product=list(product_prices.keys())
actual_price=list(product_prices.values())
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(our_line,our_line/2)
ax[0,1].bar(actual_product,actual_price,color=['maroon','green','blue','yellow'])
ax[1,0].scatter(np.random.randn(100),np.random.randn(100))
ax[1,1].hist(our_dots,color='green')
ax[0,0].set_title('Line Plot Example')
ax[0,1].set_title('Bar Chart Example')
ax[1,0].set_title('Scatter Example')
ax[1,1].set_title('Histogram Example')
ax[0,0].grid(True)
ax[1,1].grid(True)
fig.suptitle('Matplotlib Example Using 2 Rows and 2 Colums')
plt.tight_layout()
plt.savefig('SubplotExample,jpeg',dpi=300,bbox_inches='tight')
```
