```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('netflix_data.csv')
df=df.dropna(subset=['release_year'])
df['release_yearr']=df['release_year'].astype(int)
plt.figure(figsize=(8,5))
plt.hist(df['release_year'],bins=100,color='black', edgecolor='white', alpha=0.8)
plt.title('Netflix Release Title By Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('netflix_data.csv')
df=df.dropna(subset=['release_year','type'])
movies=df[df['type']=='Movie']
shows=df[df['type']=='TV Show']

plt.figure(figsize=(8,5))
plt.hist([movies['release_year'], shows['release_year']],bins=100,color=['darkred', 'navy'],edgecolor='white',alpha=0.8,label=['Movies', 'TV Shows'])

plt.title('Netflix Release Title By Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.legend()
plt.show()
```

# Use netflix_data.csv dataset
## print how many Movies and Tv Shows are available on Netflix
## Display Movies and TV shows proportion in a pie plot
## Save file as pdf

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('netflix_data.csv')
df=df.dropna(subset=['type'])
type_counts=df['type'].value_counts()
print(type_counts)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('netflix_data.csv')
df=df.dropna(subset=['type'])
type_counts=df['type'].value_counts()
print(type_counts)

plt.figure(figsize=(7,7))
plt.pie(type_counts,labels=type_counts.index,colors=['blue','maroon'],autopct='%1.2f%%',startangle=140,explode=(0.05,0),shadow=True)
plt.title('Netflix Movies and TV Shows Distribution')
plt.tight_layout()
plt.legend(loc='lower left',title='Movies and TV Shows')
# To save file
plt.savefig('Netflix Data.pdf',dpi=300,bbox_inches='tight')
plt.show()
```
