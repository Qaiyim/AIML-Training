```python
import matplotlib.pyplot as plt
```

```python
months=['Jan','Feb','March','Apr','May']
sales=[2000,3000,3500,4500,3200]
```

# plt.plot(x,y,color='color name', linestyle='line_style',linewidth=value, marker='marker_style')

```python
plt.plot(months,sales,color='pink',linestyle='-', linewidth=1, marker='o', label='Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales per Month')
plt.title('Monthly Sales Report')
plt.xlim(0,4)
plt.ylim(2000,5000)
plt.grid(True)
plt.legend(loc='lower left')
plt.show()
```

```python
import numpy as np
```

```python
x=np.linspace(0,8,100)
plt.plot(x,x,label='Linear')
plt.plot(x,x**2,label='Quadratic')
plt.plot(x,x**3, label='Cubic')
plt.xlabel('X Line')
plt.ylabel('Y Line')
plt.title('Simple Line Plot')
plt.legend(loc='lower left')
plt.show()
```

# Bar plot example
    ## plt.bar(x,y, colors='',edgecolor='')

```python
categories=['Electronics','Mobile','Books','Grocery','Home Decoration']
sales=[3400,2300,5600,3000,4000]
bar_colors=['green','yellow','red','black','blue']
```

```python
plt.title('Sales Report')
plt.bar(categories,sales,color=bar_colors, edgecolor='black')
plt.figure(figsize=(8,5))
plt.xlabel('Category')
plt.ylabel('Sales')
plt.tight_layout()
```

```python
categories=['Electronics','Mobile','Books','Grocery','Home Decoration']
sales=[58000,60000,30000,90000,40000]
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color='green', edgecolor='black')
plt.xlabel('Category')
plt.xticks(rotation=20, fontsize=12, color='magenta')
plt.title('Sales Report')
plt.ylabel('Sales')
plt.yticks([10000,20000,30000,40000,50000],['20K','40K','50K','60K','70K'],fontsize=12, color='red')
plt.tight_layout()
plt.show()
```

# Scatter plot
    ## plt.scatter(x,y,s='',color='colorname',marker, alpha=transparency, edgecolor='color name', label='label of legend')

```python
height_cm=[170,155,150,143,159]
weight_kg=[70,58,63,79,60]
```

```python
plt.figure(figsize=(8,5))
plt.scatter(height_cm,weight_kg,s=100,color='maroon', marker='*', alpha=0.9, edgecolor='black', label='weight height vomparison')
plt.title('Scatter Plot Height VS Weight')
plt.xlabel('Height CM')
plt.xticks(rotation=90, fontsize=12, color='magenta')
plt.ylabel('Weight KG')
plt.yticks([50,55,60,65,70,75,80,85],['50KG','55KG','60KG','65KG','70KG','75kg','80kg','85kg'],fontsize=12, color='red')
plt.grid(color='lightblue', alpha=0.5, linestyle='--', linewidth=0.5)
plt.show()
```
