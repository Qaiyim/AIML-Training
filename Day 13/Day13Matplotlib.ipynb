```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
temp=np.random.randint(20,100,size=50)
plt.scatter(x,y,c=temp,cmap='rainbow',s=120,edgecolor='black')
plt.colorbar(label='temperature')
plt.title('Scatter Plot Color by Temperature')
plt.xlabel('X Random')
plt.ylabel('Y Random')
plt.show()
```

```python
height_cm=[170,155,150,143,159]
weight_kg=[70,58,63,79,60]
age=[26,22,30,35,45]
```

```python
plt.figure(figsize=(8,5))
scatter=plt.scatter(height_cm,weight_kg,c=age,cmap='rainbow',s=120,alpha=0.9)
plt.colorbar(scatter,label='Age (Years)')
plt.title('Scatter Plot Height VS weight (Colored by age)')
plt.xlabel('Height CM')
plt.ylabel('Weight KG')
plt.show()
```

# Pie plot example
    ## plt.pie(x,labels='',colors='',autopct='',startangle='',shadow=,explode=,radius)

```python
department=['HR','Web Development','Banking','Editing']
expenses=[4000,5500,6500,3000]
```

```python
plt.figure(figsize=(7,7))
plt.pie(expenses,labels=department,autopct='%1.2f%%',startangle=140,explode=(0,0.05,0,0,),shadow=True,radius=1.2)
plt.title('Department Wise Expense Report')
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(expenses,labels=department,colors=plt.cm.inferno([0.6,0.3,0.5,0.7,0.9]),autopct='%1.2f%%',startangle=140,explode=(0,0.05,0,0,),shadow=True,radius=1.2)
plt.title('Department Wise Expense Report')
plt.show()
```

# Create a pie plot using following data
    ## sales={'Mobile':4500,'Clothing':13000,'Groceries':35000,'Home Decor':20000}
    ## sgow percentage with two decimal 
    ## Use colormap to give the different color to your pie

```python
sales={'Mobile':4500,'Clothing':20000,'Groceries':35000,'Home Decor':30000}
products=list(sales.keys())
sales_by=list(sales.values())
```

```python
plt.figure(figsize=(7,7))
plt.pie(sales_by,labels=products,colors=plt.cm.rainbow([0.6,0.3,0.5,0.7,0.9]),textprops={'color':'red','fontsize':10,'fontweight':'bold'}
, autopct='%1.2f%%',startangle=140,explode=(0,0.05,0,0,),shadow=True,radius=1.2)
plt.title('Sales Report')
plt.savefig('categorieswisesale.png',dpi=300,bbox_inches='tight')
plt.show()
```

# Histogram Example

```python
marks=[45,67,89,90,67,90,67,87,85,45,62,45,78,56,89,67,35,34,56,87,42]
plt.style.use('ggplot')
plt.figure(figsize=(8,5))
plt.hist(marks,bins=5,color='maroon', edgecolor='white', alpha=0.8)
plt.title('Histogram For Students Marks')
plt.xlabel('Marks Range')
plt.ylabel('Number of students')
plt.show()
```
